```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def f(x):
    return x**2+4*x+5
def df(x):
    return 2*x+4
```

```python
#梯度下降法
def gradient(s,df,p,lr):
    xr=np.zeros(p+1)
    x=s
    xr[0]=x
    for i in range(p):
        dx=df(x)
        x=x-lr*dx
        xr=x
        plt.scatter(xr,f(x))
    return xr
```

```python
plt.plot(x,f(x))
xs=random.uniform(-9,4)
times=1000
gradient(xs,df,times,0.01)
```
