```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from matplotlib import animation
```

```python
def input(x):
    y=(x-10)*(x-8)*(x-3)*(1-x)
    return y
```

```python
maxT=100
minT=1
i=1000
delta=0.95
inix=10 * (2 * np.random.rand() - 1)
nowT=maxT
a=0
print("初始解:",inix)
while nowT>minT:
    for j in range(i):
        funval=input(inix)
        newx=inix + (2 * np.random.rand() - 1)
        if -10 <= newx <= 10:
            newval=input(newx)
            deltax=newval-funval
            if deltax < 0:
                inix=newx
            else:
                p=np.exp(-deltax/nowT)
                r=np.random.rand()
                if r < p:
                    inix=newx
    nowT=nowT*delta
    a=a+1
    print(a)
print("最優值:",inix)
print("最優解:",input(inix))
```
